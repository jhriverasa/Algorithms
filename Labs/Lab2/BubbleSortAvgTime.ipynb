# Bubble Sort Average Time

This is the one implementation of Bubble Sort, receives a list(with sortable elements) and sort it. For example:


```python
def NormalBubbleSort(alist):
	for passnum in range(len(alist)-1,0,-1):
		for i in range(passnum):
			if alist[i]>alist[i+1]:
				temp = alist[i]
				alist[i] = alist[i+1]
				alist[i+1] = temp

l = [3, 5, 8, 2, 1, 7, 4, 6]
print (l) 
NormalBubbleSort(l)
print l
```

Now we're going to count total comparisons and swaps, just adding some counters in code like this, then we can return them in a list with two elements.

```python
def BubbleSort(alist):
	stats=[]
	comparisons=0
	swaps=0
	for passnum in range(len(alist)-1,0,-1):
		for i in range(passnum):
			comparisons+=1
			if alist[i]>alist[i+1]:
				swaps+=1
				temp = alist[i]
				alist[i] = alist[i+1]
				alist[i+1] = temp

	stats.append(comparisons)
	stats.append(swaps)
	return stats #stats[0] = comparisons / stats[1] = swaps

l = [3, 5, 8, 2, 1, 7, 4, 6]
print l
stats= BubbleSort(l)
print l
print "Total comparisons:"
print stats[0]
print "Total swaps:"
print stats[1]

```

In order to calculate all permutations of an n-elments set we can use itertools library.
Now, we create a function that sort (using BubbleSort with counters) every permutation generated by itertools.permutations, finally compute average comparisons and swaps.

```python
import itertools 
#Return average comparisons and swaps from sorting all permutations of 
#an  defined Array
def GetAverageStats(Array):

	n=len(Array)
	perms =list(itertools.permutations(Array))
	np = len(perms)
	avgcomparisons=0
	avgswaps=0
	for i in range(np):
		lst= list(perms[i])
		stats=BubbleSort( lst)
		avgcomparisons += stats[0]
		avgswaps += stats[1]
	avgcomparisons /= float(np)
	avgswaps /= float(np)
	r=[]
	r.append(avgcomparisons)
	r.append(avgswaps)
	return r

l=[1,3,2]
print "All permutations of [1.2,3]"
lpermutations = list(itertools.permutations(l))
print lpermutations
avgstats= GetAverageStats(l)
print "Average comparisons"
print avgstats[0]
print "Average swaps"
print avgstats[1]
```

In the example shown above, we can see that, BubbleSort requires 3 comparisons and 1.5 swaps on average. Let us show average comparisons and swaps for all permutations of 1,2,3,4,5,6,7,8 elements sets.

```python
l= []
n=8
for i in range(n):
	l.append(i+1)
	k= GetAverageStats(l)
	print ("n="+str(i+1))
	print "Average comparisons:"
	print k[0]
	print "Average swaps:"
	print k[1]
	print "--------------------"
```
